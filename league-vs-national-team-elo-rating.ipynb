```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('elo.xlsx', index_col=0)
df.head(5)
```

```python
plt.figure(figsize = (15, 15), dpi = 200, facecolor = '#f4f2e5')

ax = plt.axes()
ax.set_facecolor("#f4f2e5")

ax.grid(True, c = "grey", ls = ":")

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.scatter(df['Team Elo'], df['League Elo'], s = 200, c = "lightblue", ec = "blue", alpha = 0.5)

for i in range(1, len(df)+1):
    if df['Code'][i] == "GER":
        plt.text(df['Team Elo'][i] - 50, df['League Elo'][i] + 10, s = df['Code'][i], fontsize=15)
    elif df['Code'][i] == "BEL":
        plt.text(df['Team Elo'][i] - 50, df['League Elo'][i] + 10, s = df['Code'][i], fontsize=15)
    elif df['Code'][i] == "AZE":
        plt.text(df['Team Elo'][i] - 50, df['League Elo'][i] + 10, s = df['Code'][i], fontsize=15)
    elif df['Code'][i] == "MKD":
        plt.text(df['Team Elo'][i] - 70, df['League Elo'][i] + 10, s = df['Code'][i], fontsize=15)
    elif df['Code'][i] == "GEO":
        plt.text(df['Team Elo'][i] - 50, df['League Elo'][i] + 10, s = df['Code'][i], fontsize=15)
    elif df['Code'][i] == "GRE":
        plt.text(df['Team Elo'][i] - 50, df['League Elo'][i] + 10, s = df['Code'][i], fontsize=15)
    elif df['Code'][i] == "AUT":
        plt.text(df['Team Elo'][i] - 30, df['League Elo'][i] + 10, s = df['Code'][i], fontsize=15)    
    elif df['Code'][i] == "NOR":
        plt.text(df['Team Elo'][i] + 10, df['League Elo'][i] - 35, s = df['Code'][i], fontsize=15)
    elif df['Code'][i] == "SCO":
        plt.text(df['Team Elo'][i] + 10, df['League Elo'][i] - 15, s = df['Code'][i], fontsize=15)
    elif df['Code'][i] == "FIN":
        plt.text(df['Team Elo'][i] + 10, df['League Elo'][i] - 10, s = df['Code'][i], fontsize=15)
    else:
        plt.text(df['Team Elo'][i] + 10, df['League Elo'][i] + 10, s = df['Code'][i], fontsize=15)
        
plt.axvline(df['Team Elo'].mean(), linewidth = 3, ls = ":")
plt.axhline(df['League Elo'].mean(), linewidth = 3, ls = ":")

plt.title("National Team vs National League Elo Comparison", fontsize = 20)

plt.xlabel("National Team Elo Rating", fontsize = 15)
plt.ylabel("League Top 10 Avg Elo Rating", fontsize = 15)

plt.text(780, 1550, "Average League Elo", fontsize = 10)
plt.text(1605, 1850, "Average National Team Elo", fontsize = 10, rotation = 90)

plt.text(1700, 750, "Data Source:", fontsize = 15)
plt.text(1700, 720, "eloratings.net & elofootball.com", fontsize = 15)

plt.savefig('transfer-balance.png')

plt.show()
```
