```python
## Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Importing dataset created in excel with data used from www.transfermarkt.com

df = pd.read_excel('totaltransferpaid.xlsx')
df.head(5)
```

```python
## Dividing clubs to Leagues

eng_df = df.loc[df['League'] == 'Premier League']
esp_df = df.loc[df['League'] == 'La Liga']
ita_df = df.loc[df['League'] == 'Serie A']
ger_df = df.loc[df['League'] == 'Bundesliga']
fra_df = df.loc[df['League'] == 'Ligue 1']
cl_df = df.loc[df['League'] == 'Champions League']
```

```python
## Creating visualisation

plt.figure(figsize = (15,15), dpi = 200, facecolor = '#f4f2e5')

ax = plt.axes()
ax.set_facecolor("#f4f2e5")

ax.grid(True, c = "grey", ls = ":")

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.scatter(eng_df['Total Inc'], eng_df['Total Exp'], s = 100, c = "#00ff85", ec = "black", alpha = 0.5, label = "Premier League")
plt.scatter(esp_df['Total Inc'], esp_df['Total Exp'], s = 100, c = "#ee8707", ec = "black", alpha = 0.5, label = "La Liga")
plt.scatter(ita_df['Total Inc'], ita_df['Total Exp'], s = 100, c = "#008fd7", ec = "black", alpha = 0.5, label = "Serie A")
plt.scatter(ger_df['Total Inc'], ger_df['Total Exp'], s = 100, c = "#d20515", ec = "black", alpha = 0.5, label = "Bundesliga")
plt.scatter(fra_df['Total Inc'], fra_df['Total Exp'], s = 100, c = "#dae025", ec = "black", alpha = 0.5, label = "Ligue 1")
plt.scatter(cl_df['Total Inc'], cl_df['Total Exp'], s = 100, c = "#0e1e5b", ec = "black", alpha = 0.5, label = "Champions League")

## Adding names to data points and arranging overlapping text 

for i in range(0, len(df)):
    if df['ShortName'][i] == 'Milan':
        plt.text(df['Total Inc'][i] + 5, df['Total Exp'][i] - 5, s = df['ShortName'][i])
    elif df['ShortName'][i] == 'Frankfurt':
        plt.text(df['Total Inc'][i], df['Total Exp'][i] - 15, s = df['ShortName'][i])
    elif df['ShortName'][i] == 'West Ham':
        plt.text(df['Total Inc'][i], df['Total Exp'][i] + 15, s = df['ShortName'][i])
    elif df['ShortName'][i] == 'A. Villa':
        plt.text(df['Total Inc'][i] - 20, df['Total Exp'][i] + 5, s = df['ShortName'][i])
    elif df['ShortName'][i] == 'Brighton':
        plt.text(df['Total Inc'][i] - 20, df['Total Exp'][i] + 5, s = df['ShortName'][i])
    elif df['ShortName'][i] == 'Nice':
        plt.text(df['Total Inc'][i] - 20, df['Total Exp'][i] + 5, s = df['ShortName'][i])
    elif df['ShortName'][i] == 'Hertha':
        plt.text(df['Total Inc'][i] - 35, df['Total Exp'][i] - 10, s = df['ShortName'][i])
    elif df['ShortName'][i] == 'Benfica':
        plt.text(df['Total Inc'][i] - 20, df['Total Exp'][i] + 10, s = df['ShortName'][i])
    elif df['ShortName'][i] == 'Soton':
        plt.text(df['Total Inc'][i] - 20, df['Total Exp'][i] + 5, s = df['ShortName'][i]) 
    elif df['ShortName'][i] == 'Valencia':
        plt.text(df['Total Inc'][i] + 5, df['Total Exp'][i] - 15, s = df['ShortName'][i])
    elif df['ShortName'][i] == 'Hoffenheim':
        plt.text(df['Total Inc'][i] + 5, df['Total Exp'][i] - 15, s = df['ShortName'][i])
    elif df['ShortName'][i] == 'Shakhtar':
        plt.text(df['Total Inc'][i], df['Total Exp'][i], s = "")
    elif df['ShortName'][i] == 'Sociedad':
        plt.text(df['Total Inc'][i], df['Total Exp'][i], s = "")
    elif df['ShortName'][i] == 'Betis':
        plt.text(df['Total Inc'][i], df['Total Exp'][i], s = "")
    elif df['ShortName'][i] == 'Torino':
        plt.text(df['Total Inc'][i], df['Total Exp'][i], s = "")
    elif df['ShortName'][i] == 'Brugge':
        plt.text(df['Total Inc'][i], df['Total Exp'][i], s = "")
    elif df['Total Exp'][i] > 200:
        plt.text(df['Total Inc'][i]+5, df['Total Exp'][i]+5, s = df['ShortName'][i])
    elif df['Total Inc'][i] > 120:
        plt.text(df['Total Inc'][i]+5, df['Total Exp'][i]+5, s = df['ShortName'][i])
        
plt.title("Transfer Balance - Last 5 seasons - Top European Clubs", fontsize = 20)

plt.xlabel("Total Income", fontsize = 15)
plt.ylabel("Total Expenditure", fontsize = 15)

p = np.poly1d(np.polyfit(df['Total Inc'], df['Total Exp'], 1))
plt.plot(df['Total Inc'], p(df['Total Inc']), ":", c = "black", alpha = 0.5)

plt.legend(facecolor = "#f4f2e5")

plt.text(450, -20, "Data Source: Transfermarkt.com", fontsize = 15)

plt.savefig('transfer-balance.png')

plt.show()
```
