```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame()
```

```python
for i in range(2015, 2023):
    season = f"{i}-{i+1}"
    site = f"https://fbref.com/en/comps/Big5/{season}/shooting/players/{season}-Big-5-European-Leagues-Stats"
    page = pd.read_html(site)
    frame = page[0]
    frame.columns = frame.columns.droplevel(0)
    frame = frame[frame['Gls'].notna()]
    frame = frame[frame['Gls'] != 'Gls']
    frame['Gls'] = frame['Gls'].astype(int)
    frame['Season'] = i+1
    
    df = pd.concat([df, frame])
```

```python
goals = df[['Player', 'Nation', 'Pos', 'Squad', 'Comp', 'Age', '90s', 'Gls', 'PK', 'xG', 'npxG']]
goals = goals[goals['xG'].notna()]
```

```python
goals['npGls'] = goals['Gls'] - goals['PK'].astype(float)
goals['npxG'] = goals['npxG'].astype(float)
```

```python
goals = goals.groupby(['Player'], as_index=False).sum()

goals['Diff'] = goals['npGls'] - goals['npxG']
goals = goals.sort_values(by=['Diff'])
```
